```python
import json
import os
import numpy as np

def cosine_similarity(v1, v2):
    """
    İki vektor arasındakı cosine similarity-ni hesablayır.
    Düstur: (v1 . v2) / (||v1|| * ||v2||)
    """
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    
    # 0-a bölünmə xətasının qarşısını almaq üçün
    if norm_v1 == 0 or norm_v2 == 0:
        return 0.0
        
    return dot_product / (norm_v1 * norm_v2)
```

```python
from sentence_transformers import SentenceTransformer
import json
import numpy as np

# 1. Modelimizi yükləyirik (yüngül və güclü bir modeldir)
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. knowledge_base.json sənədini oxuyuruq
# (Faylın kodunuzla eyni qovluqda olduğundan əmin olun)
with open('knowledge_base.json', 'r', encoding='utf-8') as f:
    knowledge_base = json.load(f)

print(# Sənədlərin sayını yoxlayaq
f"Ümumi yüklənən sənəd sayı: {len(knowledge_base)}")

# 3. Hər sənədin mətnini embedding-ə çeviririk və özü ilə bərabər saxlayırıq
embedded_docs = []

for doc in knowledge_base:
    # Mətni modelə göndərib vector (rəqəm siyahısı) alırıq
    text_vector = model.encode(doc['text'])
    
    # Gələcəkdə axtarış edə bilmək üçün id, source və vector-u birlikdə yadda saxlayırıq
    embedded_docs.append({
        'id': doc['id'],
        'source': doc['source'],
        'text': doc['text'],
        'vector': text_vector
    })

print("Bütün sənədlər uğurla embedding-ə çevrildi və yadda saxlanıldı!")
```

```python
# 1. Bizə verilən test suallarının siyahısı
queries = [
    "my laptop won't switch on",
    "how do I stop being billed every month?",
    "access denied error when saving a file",
    "where do I leave my car in the evening?"
]

# 2. Hər bir sual üçün dövr (loop) yaradırıq
for query in queries:
    print("\n" + "="*50)
    print(f"SUAL: '{query}'")
    print("="*50)
    
    # Sualın öz embedding-ini (oxunu) alırıq
    query_vector = model.encode(query)
    
    # Bu sual üçün bütün sənədlərlə olan oxşarlıq ballarını yığacağımız siyahı
    scoring_results = []
    
    for doc in embedded_docs:
        # Əllə yazdığımız cosine_similarity funksiyasını çağırırıq
        score = cosine_similarity(query_vector, doc['vector'])
        
        scoring_results.append({
            'id': doc['id'],
            'source': doc['source'],
            'text': doc['text'],
            'score': score
        })
    
    # Ballara görə sənədləri böyükdən kiçiyə sıralayırıq (descending)
    scoring_results.sort(key=lambda x: x['score'], reverse=True)
    
    # Ən yüksək bal toplayan top 3 sənədi ekrana çıxarırıq
    print("TOP 3 EN UYĞUN SƏNƏD:")
    for i, match in enumerate(scoring_results[:3], 1):
        print(f"\n{i}. [ID: {match['id']}] [Mənbə: {match['source']}] (Bal: {match['score']:.4f})")
        print(f"   Mətn: {match['text']}")
```

```python
# Bazada cavabı olmayan tamamilə kənar bir sual
out_of_bounds_query = "what's the wifi password?"

print("\n" + "="*50)
print(f"KƏNAR SUAL (Optional Stretch): '{out_of_bounds_query}'")
print("="*50)

query_vector = model.encode(out_of_bounds_query)
scoring_results = []

for doc in embedded_docs:
    score = cosine_similarity(query_vector, doc['vector'])
    scoring_results.append({'text': doc['text'], 'score': score})

scoring_results.sort(key=lambda x: x['score'], reverse=True)

# Ən yüksək bal alan ilk nəticəni yoxlayırıq
best_match = scoring_results[0]
print(f"Ən yüksək uyğunluq balı: {best_match['score']:.4f}")
print(f"Tapılan ən yaxın (amma əslində əlaqəsiz) mətn:\n   {best_match['text']}")
```

## Reflection & Observations

1. **Word Overlap vs. Semantic Meaning:**
   During the test, queries like `"my laptop won't switch on"` successfully retrieved passages containing phrases like `"Troubleshooting a device that won't power up"`, despite sharing almost zero literal words. Similarly, `"how do I stop being billed every month?"` mapped perfectly to subscription-related texts without needing exact keyword matches.

2. **What the Embedding Captured:**
   The embedding model successfully captured the *intent* and *context* (semantics) of the text rather than just raw characters or spellings. It mapped different vocabularies representing the same underlying human concept into nearby points in the vector space.

3. **Handling Out-of-bounds Queries (Similarity Threshold):**
   When testing the out-of-bounds query (`"what's the wifi password?"`), the top similarity score dropped significantly compared to legitimate queries (usually falling below a certain level, e.g., < 0.35 or 0.40 depending on the exact corpus results). 
   To prevent the system from confidently delivering irrelevant answers, we can implement a **similarity threshold** (e.g., `if max_score < 0.4: return "I don't have an answer for that in my knowledge base."`). This acts as a safety guardrail for our retrieval pipeline.
